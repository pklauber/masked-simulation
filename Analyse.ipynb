```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This is also doable with a List of Lists, but that reduces the code readability by A LOT so I decided to go with this less efficient but easier-to-understand approach. Its a py-Notebook after all.
#Mask-Percentage = 5%
e55 = pd.read_csv("Experiment_data/experiment-5-5.csv")
e520 = pd.read_csv("Experiment_data/experiment-5-20.csv")
e530 = pd.read_csv("Experiment_data/experiment-5-30.csv")
e550 = pd.read_csv("Experiment_data/experiment-5-50.csv")
e565 = pd.read_csv("Experiment_data/experiment-5-65.csv")

#Mask-Percentage = 20%
e205 = pd.read_csv("Experiment_data/experiment-20-5.csv")
e2020 = pd.read_csv("Experiment_data/experiment-20-20.csv")
e2030 = pd.read_csv("Experiment_data/experiment-20-30.csv")
e2050 = pd.read_csv("Experiment_data/experiment-20-50.csv")
e2065 = pd.read_csv("Experiment_data/experiment-20-65.csv")

#Mask-Percentage = 50%
e505 = pd.read_csv("Experiment_data/experiment-50-5.csv")
e5020 = pd.read_csv("Experiment_data/experiment-50-20.csv")
e5030 = pd.read_csv("Experiment_data/experiment-50-30.csv")
e5050 = pd.read_csv("Experiment_data/experiment-50-50.csv")
e5065 = pd.read_csv("Experiment_data/experiment-50-65.csv")

#Mask-Percentage = 70%
e705 = pd.read_csv("Experiment_data/experiment-70-5.csv")
e7020 = pd.read_csv("Experiment_data/experiment-70-20.csv")
e7030 = pd.read_csv("Experiment_data/experiment-70-30.csv")
e7050 = pd.read_csv("Experiment_data/experiment-70-50.csv")
e7065 = pd.read_csv("Experiment_data/experiment-70-65.csv")

#Mask-Percentage = 100%
e1005 = pd.read_csv("Experiment_data/experiment-100-5.csv")
e10020 = pd.read_csv("Experiment_data/experiment-100-20.csv")
e10030 = pd.read_csv("Experiment_data/experiment-100-30.csv")
e10050 = pd.read_csv("Experiment_data/experiment-100-50.csv")
e10065 = pd.read_csv("Experiment_data/experiment-100-65.csv")
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)

plt.plot(e55["infected turtles"], label = "5%")
plt.plot(e205["infected turtles"], label = "20%")
plt.plot(e505["infected turtles"], label = "50%")
plt.plot(e705["infected turtles"], label = "70%")
plt.plot(e1005["infected turtles"], label = "100%")

plt.title('Infected turtles over time with mask protection = 5%')

plt.xlabel("Time")
plt.ylabel("Number of turtles")

plt.legend()
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)

plt.plot(e705["infected turtles"], label = "infected")
plt.plot(e705["exposed turtles"], label = "exposed")
plt.plot(e705["immune turtles"], label = "immune")

plt.title('Relation of Exposed to infected to immune turtles with 70% Mask Percentage and 5% Mask Protection')

plt.xlabel("Time")
plt.ylabel("Number of turtles")

plt.legend()
plt.show()
```
